```python
#We want to see how Mexican Economy has evolved over time. We will analyse 
#Mexican exports and identify trade patterns have evolved.
#The information was downloaded from the WTO web page from 1980 to 2016.

import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import seaborn as sns
import pandas as pd
import csv

df = pd.read_csv('MexExpImp.csv')
```

```python
#Total exports
array1 = ["Exports"]
total=df.loc[df['Flow_desc'].isin(array1)]
total1 = total.groupby(['Year'])['Value'].sum().reset_index(name ='Value')

total1

plt.subplots(figsize=(7,7))
sns.barplot(x = "Value", y = "Year", orient="h", data = total1, palette='rainbow')
```

```python
array2 = ["1980", "1994", "2016"]
nafta=total.loc[total['Year'].isin(array2)]
nafta


grupo2 = nafta.groupby(["Year", "Indicator_desc"])['Value'].sum().reset_index(name ='Value')
plt.subplots(figsize=(17,17))
graf2=sns.barplot(x="Year", y="Value", hue="Indicator_desc", data=grupo2)
graf2.set_xticklabels(graf2.get_xticklabels(),rotation=90)
```

```python
#As we can observe, the Mexican exports have grown considerably in almost every sector.
```

```python
array3 = [" Machinery and transport equipment", "  Transport equipment", "   Automotive products"]
auto=total.loc[total['Indicator_desc'].isin(array3)]
auto


grupo3 = auto.groupby(["Year", "Indicator_desc"])['Value'].sum().reset_index(name ='Value')
plt.subplots(figsize=(10,10))
graf3=sns.barplot(x="Year", y="Value", hue="Indicator_desc", data=grupo3)
graf3.set_xticklabels(graf3.get_xticklabels(),rotation=90)
```

```python
#Exports related to the automotive industry have grown constantly through time. We may clearly 
#identify 2009 as a low export year for Mexico since we had an intenational crisis.
```

```python
array4 = ["  Office and telecom equipment", "   Electronic data processing and office equipment", "   Telecommunications equipment", "   Integrated circuits and electronic components"]
elec=total.loc[total['Indicator_desc'].isin(array4)]
elec


grupo4 = elec.groupby(["Year", "Indicator_desc"])['Value'].sum().reset_index(name ='Value')
plt.subplots(figsize=(10,10))
graf4=sns.barplot(x="Year", y="Value", hue="Indicator_desc", data=grupo4)
graf4.set_xticklabels(graf4.get_xticklabels(),rotation=90)
```

```python
#We can see the same pattern as in the automotive industry. Despite a strong growht over the
#period, during 2016 Mexican exports slowed down.
```
